```python
#1
from datetime import datetime
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_1.csv")
df["result"] = pd.to_datetime(df["도착시간"]) - pd.to_datetime(df["출동시간"])
df["result"] = df["result"].astype(str)
df["result"] = df["result"].str.replace('0 days ', '')
df[["시간","분","초"]] = df["result"].str.split(':',expand=True)
df["시간"] = df["시간"].astype(int)
df["분"] = df["분"].astype(int)
df["초"] = df["초"].astype(int)
df["result2"] = df["시간"]*60 + df["분"] + (df["초"] / 60)
mean = df.groupby("구급대")["result2"].mean() #구급대별로 구해야 함
print(round(mean.max()))
```

```python
#2
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_2.csv")

df.set_index("대학교명", inplace=True)

df["student_num"] = df.loc[:, df.columns!="교수"].sum(axis=1) / df["교수"]
d_max = df["student_num"].max()

result = df[df["student_num"] == d_max][["교수"]]
print(result.iloc[0][0])
```

```python
#3
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_3.csv")

df[['년', '월']] = df['날짜'].str.split(' ', expand=True)
df["합"] = df.loc[:,"폭력범죄":"정치범죄"].sum(axis=1)
print(round(df.groupby("년")["합"].mean().max())) #년도별로 구해야 함
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score

df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_train.csv")

df["키"] = LabelEncoder().fit_transform(df["키"]) # string 변환
df["체격"] =  LabelEncoder().fit_transform(df["체격"]) # string 변환

X = df.loc[:, "일평균수면시간":"기초대사량"]
y = LabelEncoder().fit_transform(df["구분"]) #string 변환

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=11)
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train,y_train)

pred = dt.predict(X_test)
print(accuracy_score(y_test,pred))

test_df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_test.csv")
test_df["키"] = LabelEncoder().fit_transform(test_df["키"])
test_df["체격"] = LabelEncoder().fit_transform(test_df["체격"])
pred_test = dt.predict(test_df.loc[:,"일평균수면시간":"기초대사량"])
result = pd.DataFrame({'pred':pred_test})
result.to_csv('result.csv',index=False)
```
