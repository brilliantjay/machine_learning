제 2 유형

```python
loan = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/Bank_Personal_Loan_Modeling.csv")
loan
```

```python
loan2 = loan[loan["Personal Loan"] == 1]
loan2
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

loan.info()
```

```python
X = loan.iloc[:, loan.columns !='Personal Loan']
y = loan['Personal Loan']

X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler

X_scaled = MinMaxScaler().fit_transform(X)
X = pd.DataFrame(X_scaled, columns=X.columns)
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=11)

result_list= []

for i in range(1,20,1):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  pred = knn.predict(X_test)
  res = accuracy_score(y_test,pred)
  result_list.append((i,res))

max_acc = 0
n = 0
for r in result_list:
  a, b = r
  if b > max_acc:
    max_acc = b
    n = a
print(n)
print(max_acc)
```
