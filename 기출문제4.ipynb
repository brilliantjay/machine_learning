```python
###제2유형
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

CS_Seg_X_train = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/CS_Seg_X_train.csv")
CS_Seg_X_test = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/CS_Seg_X_test.csv")
CS_Seg_y_train = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/CS_Seg_y_train.csv")

CS_Seg_X_train["Gender"] = LabelEncoder().fit_transform(CS_Seg_X_train["Gender"])
CS_Seg_X_train["Ever_Married"] = LabelEncoder().fit_transform(CS_Seg_X_train["Ever_Married"])
CS_Seg_X_train["Graduated"] = LabelEncoder().fit_transform(CS_Seg_X_train["Graduated"])
CS_Seg_X_train["Profession"] = LabelEncoder().fit_transform(CS_Seg_X_train["Profession"])
CS_Seg_X_train["Spending_Score"] = LabelEncoder().fit_transform(CS_Seg_X_train["Spending_Score"])
CS_Seg_X_train

CS_Seg_X_test["Gender"] = LabelEncoder().fit_transform(CS_Seg_X_test["Gender"])
CS_Seg_X_test["Ever_Married"] = LabelEncoder().fit_transform(CS_Seg_X_test["Ever_Married"])
CS_Seg_X_test["Graduated"] = LabelEncoder().fit_transform(CS_Seg_X_test["Graduated"])
CS_Seg_X_test["Profession"] = LabelEncoder().fit_transform(CS_Seg_X_test["Profession"])
CS_Seg_X_test["Spending_Score"] = LabelEncoder().fit_transform(CS_Seg_X_test["Spending_Score"])
CS_Seg_X_test

```

```python
CS_Seg_y_train
```

```python
from sklearn.ensemble import RandomForestClassifier

X_train,X_test,y_train,y_test = train_test_split(CS_Seg_X_train,CS_Seg_y_train["Segmentation"].values,test_size=0.2,random_state=11)

rf = RandomForestClassifier(n_estimators=1000,max_depth=10,random_state=20)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
print(accuracy_score(y_test,pred))


```
